```python
#climate grid creation 
import xarray as xr
temp_grid = xr.DataArray(
    [[30,31,32],
     [28,29,30]],

     dims= ["lat","lon"],
     coords={
         "lat": [30,31],
         "lon": [75,76,77]
     }
)
temp_grid
```

```python
print(temp_grid)
print(temp_grid.values)
print(temp_grid.dims)
```
