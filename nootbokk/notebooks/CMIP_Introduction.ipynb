```python
import xarray as xr

ds = xr.Dataset(
    data_vars={
        "temperature": (
            ["time", "lat", "lon"],
            [[[30, 31],
              [29, 28]],
             
             [[32, 33],
              [30, 29]]]
        )
    },
    coords={
        "time": ["2025-01-01", "2025-01-02"],
        "lat": [30, 31],
        "lon": [75, 76]
    }
)

ds
```

```python
ds.to_netcdf("my_first_climate_dataset.nc")
```
